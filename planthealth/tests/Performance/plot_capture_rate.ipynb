```python
delay = np.array([.0125,.025,.05,0.1,0.15,0.2,0.25,0.3,0.35,.4])
frames = np.array([0,0,0,0,0,6,21,25,53,0]) #baudrate 100000
frames2 = np.array([0,0,0,0,0,29,55,86,125,300]) #baudrate 10000
frames3 = np.array([27,31,29,63,300,300,300,300,300,300]) #baudrate 7500
```

```python
plt.figure(figsize=(7,5))
plt.stem(delay*1000+1, frames, 'xk')
plt.stem(delay*1000+1, frames2, 'xr')
plt.stem(delay*1000+1, frames3, 'xy')
plt.title('Captures Before I/O Error')
plt.xlabel('delay (ms)')
plt.ylabel('frames captured')
plt.legend(['baudrate=100000','baudrate=10000','baudrate=7500'])
plt.show()
```
